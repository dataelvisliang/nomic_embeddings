```python
from nomic import embed
```

```python
output = embed.text(
    texts=["dog", "cat", "puppy", "kitten",
    "car", "truck", "vehicle", "automobile",
    "apple", "orange", "banana", "fruit"],
    model='nomic-embed-text-v1.5',
    task_type='clustering',
    inference_mode='local',
    device='cuda'  # or 'cuda' for GPU
)

print(output['embeddings'])
print(f"Shape: {len(output['embeddings'])} texts, {len(output['embeddings'][0])} dimensions")
```
